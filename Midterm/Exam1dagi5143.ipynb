#  Exam 1 (110pts) 

The following dictionary shows bakeries, their prices for cupcakes, cookies, and donuts, and the daily average profit

```python
bak = {
        'Sprinkles Bakery': ['cupcakes', 2.99, 'cookies', 0.99, 'donuts', 1.98, 339.59],
        'Guilty Pleasures': ['cupcakes', 3.99, 'cookies', 1.48, 'donuts', 2.99, 250.78],
        'Sugarlips Cupcakes': ['cupcakes', 2.99, 'cookies', 1.59, 'donuts', 1.59, 123.89],
        'Holy Cupcakes!': ['cupcakes', 2.59, 'cookies', 2.99, 'donuts', 0.59, 300.00],
        'Whipped': ['cupcakes', 3.50, 'cookies', 1.00, 'donuts', 1.50, 145.89],
        'Sugar Mama\'s Cupcakes': ['cupcakes', 3.78, 'cookies', 0.98,  'donuts', 0.89, 400.98],
        'Frankencake': ['cupcakes', 3.00, 'cookies', 1.69, 'donuts', 1.29, 666.66]
        }
```

## 1 (10 pts)
Print the bakery name, the total price (rounded to 2 decimal) for all their products,  
and "cheap" if the total price for cupcakes, cookies and donuts is under 6.00 otherwise print "expensive" 

```python
for x in bak.keys():
    total = bak[x][1] + bak[x][3] + bak[x][5]
    if total < 6:
        print(x, round(total,2), 'Cheap')
    else:
        print(x, total, 'Expensive')
```

## 2 (10pts)
Determine the average overall daily profit (round to 2 decimals) for all bakeries<br>
Make sure the calculation is still correct when a bakery is added

```python
total = 0
for k in bak.keys():
    profit = bak[k][-1]
    total += profit
    length = len(bak.keys())
    avgprofit = total/length
print(round(avgprofit,2))
 
```
