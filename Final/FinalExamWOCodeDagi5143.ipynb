## 5) Create a pivot table that shows the total sales by artist and within the artist by album and within album by track. Split the revenue by media type. Fill missing values with zero. Create an overall total. (40 pts)

```python
tracalb = tracks.merge(albums, left_on='albumid', right_on='albumid')
tracalb.head(3)
```

```python
allthree = tracalb.merge(sales, left_on='trackid', right_on='trackid')
allthree.head(3)
```

```python
allthree['revenue'] = allthree['quantity'] * allthree['unitPrice']
```

```python
thisthing = pd.pivot_table(allthree, index=['artist', 'title', 'name'], values=['revenue'], columns=['mediaType'], aggfunc=[np.sum], fill_value = 0, margins = True)
```

```python
thisthing
```
